# Defining Crawling Data by Place

```python
# reference: https://medium.com/analytics-vidhya/how-to-scrape-tweets-and-create-dataset-using-twint-without-twitter-api-e5890c25d1c9

import twint
import nest_asyncio
nest_asyncio.apply()
from datetime import datetime
import os

def scrape_by_city(keywords, folders, cities):
    os.
    if folders[len(folders) - 1] != '/':
        folders = folders + '/'
    for city in cities:
        print(city)
        c = twint.Config()
        final_keywords = keywords + ' -filter:links -filter:retweets near:"{}"'.format(city)
        c.Search = final_keywords #search keyword
        c.Since = '2021-01-19'
        c.Until = '2021-07-19'
        c.Store_csv = True
        c.Custom['tweet'] = ['id','user_id', 'tweet']
        c.Output = folders + "{}.csv".format(final_keywords)
        c.Hide_output = False
        c.Limit = 50000
        twint.run.Search(c)
```

```python
QUERY = "(covid OR corona OR covid19 OR covid-19 OR ) (vaksinasi OR vaksin OR vaksinisasi) (\"berguna\" OR berhasil OR menghidupkan OR membantu OR bantu)" #@param {type:"string"}
OUT_DIRECTORY = '/content/drive/Shareddrives/GEMASTIK XIV: Yaudahlah/data/validation/raw/berhasil' #@param {type:"string"}
place = ['DKI JAKARTA', 'JAWA BARAT', 'JAWA TENGAH', 'DAERAH ISTIMEWA YOGYAKARTA', 'JAWA TIMUR', 'BANTEN'] #@param {type:"raw"}

scrape_by_city(QUERY, OUT_DIRECTORY, place)
```

# Defining Crawling Data Normal

```python
import os
import twint
import nest_asyncio
nest_asyncio.apply()

def scrape(keywords, folders):
  if folders[len(folders) - 1] != '/':
        folders = folders + '/'
  if not os.path.exists(folders):
    print("Creating Directory {}".format(folders))
    os.mkdir(folders)
  c = twint.Config()
  final_keywords = keywords + ' -filter:links -filter:retweets -filter:replies lang:id'
  c.Search = final_keywords
  c.Since = '2021-01-19'
  c.Until = '2021-07-19'
  c.Store_csv = True
  c.Custom['tweet'] = ['id','user_id', 'tweet']
  c.Output = folders + "{}.csv".format(final_keywords)
  c.Hide_output = False
  c.Limit = 30000
  twint.run.Search(c)
```

```python
QUERY = "(covid OR corona OR covid19 OR covid-19 OR ) (vaksinasi OR vaksin OR vaksinisasi) sinopharm" #@param {type:"string"}
OUT_DIRECTORY = '/content/drive/Shareddrives/GEMASTIK XIV: Yaudahlah/data/validation/raw' #@param {type:"string"}

scrape(QUERY, OUT_DIRECTORY)
```
